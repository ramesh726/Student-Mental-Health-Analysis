```python
def classify_stress(row):
    if row['Stress Level'] == 'High' and row['Anxious Before Exams'] == 'Yes':
        return 'High Risk'
    elif row['Stress Level'] == 'Medium' or row['Anxious Before Exams'] == 'Yes':
        return 'Moderate Risk'
    else:
        return 'Low Risk'

data.loc[:, 'Stress Risk Category'] = data.apply(classify_stress, axis=1)
```

```python
sns.violinplot(x=data["Stress Level"], y=data["Sleep Duration (hrs)"])
plt.title("Sleep Duration Distribution by Stress Level")
plt.xlabel("Stress Level")
plt.ylabel("Hours of Sleep")
plt.show()
```

```python
sns.violinplot(data=data[["Screen Time (hrs/day)", "Sleep Duration (hrs)"]])
plt.title("Comparison of Screen Time vs Sleep Duration")
plt.xlabel("Variable")
plt.ylabel("Hours")
plt.xticks(ticks=[0,1], labels=["Screen Time", "Sleep Duration"])
plt.show()
```

```python
sns.jointplot(x=data["Screen Time (hrs/day)"], y=data["Sleep Duration (hrs)"], kind="scatter")
plt.suptitle("Screen Time vs Sleep Duration", y=1.02)
plt.show()
```

```python
#A regression line is used in statistical analysis to show the relationship between two variables and predict future values
sns.jointplot(x=data["Screen Time (hrs/day)"], y=data["Sleep Duration (hrs)"], kind="reg")
plt.suptitle("Regression Plot: Screen Time vs Sleep Duration", y=1.02)
plt.show()
```

```python
# The y=1.02 inside plt.suptitle(...) controls the vertical positioning of the title above the pair plot.
sns.pairplot(data, hue="Academic Performance Change")
plt.suptitle("Pairplot Colored by Academic Performance", y=1.02)
plt.show()
```

```python
## A scatter plot is used to visualize the relationship between two variables by plotting individual data points.

sns.scatterplot(x=data["Screen Time (hrs/day)"], y=data["Anxious Before Exams"])
plt.title("Screen Time vs Sleep Duration")
plt.xlabel("Screen Time (hrs/day)")
plt.ylabel("Anxious Before Exams")
plt.show()
```

```python
# A violin plot is a combination of a box plot and a kernel density plot. It helps visualize both the distribution (spread) and density (concentration) of a dataset.
sns.violinplot(x=data["Gender"], y=data["Screen Time (hrs/day)"])
plt.title("Screen Time Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Screen Time (hrs/day)")
plt.show()
```

```python
sns.boxplot(x=data["Stress Level"], y=data["Sleep Duration (hrs)"])
plt.title("Sleep Duration Distribution Across Stress Levels")
plt.xlabel("Stress Level")
plt.ylabel("Sleep Duration (hrs)")
plt.show()
```

```python
#A swarm plot is a useful visualization for showing the distribution of individual data points while preventing overlap. It helps analyze how screen time varies across different stress levels in your dataset
sns.swarmplot(x=data["Stress Level"], y=data["Screen Time (hrs/day)"])
plt.title("Screen Time Distribution by Stress Level")
plt.xlabel("Stress Level")
plt.ylabel("Screen Time (hrs/day)")
plt.show()
```
